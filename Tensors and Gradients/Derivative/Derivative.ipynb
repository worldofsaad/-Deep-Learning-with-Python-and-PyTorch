```python
import torch
import matplotlib.pylab as plt
import torch.functional as F


# Calculate the y = x^2 + 2x + 1, then find the derivative at x=2
x = torch.tensor(2.0,requires_grad= True)
y =  x**2 + 2*x + 1
print('The function: ',y)
y.backward()
print('the derivative at x',x.grad)


# Calculate the derivative with multiple values -10 to 10 with 10 steps Y = x ** 2
x = torch.linspace(-10,10,10,requires_grad=True)
Y = torch.sin(x ** 2)
y = torch.sum(Y)
y.backward()
plt.plot(x.detach().numpy(), Y.detach().numpy(), label='function')
plt.plot(x.detach().numpy(), x.grad.numpy(), label='derivative')
plt.xlabel('x')
plt.legend()
plt.show()
```
